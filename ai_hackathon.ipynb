```python
import torch
from transformers import AutoTokenizer, AutoModel
import numpy as np

# Load the BERTweet model and tokenizer
model_name = 'vinai/bertweet-base'
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModel.from_pretrained(model_name)

# Ensure the model is on GPU if available
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model.to(device)

# Function to get embeddings in batches
def get_embeddings_in_batches(texts, batch_size=32):
    embeddings = []
    model.eval()  # Evaluation mode to disable dropout
    with torch.no_grad():  # Disable gradient calculation
        for i in range(0, len(texts), batch_size):
            batch = texts[i:i+batch_size]
            inputs = tokenizer(batch, return_tensors='pt', padding=True, truncation=True, max_length=128).to(device)  # Move inputs to device
            outputs = model(**inputs)
            # Take the mean of the token embeddings for sentence embeddings
            batch_embeddings = outputs.last_hidden_state.mean(dim=1).cpu().numpy()  # Move the output back to CPU
            embeddings.append(batch_embeddings)
    return np.vstack(embeddings)

# Example usage
texts = df['combined_text'].tolist()  # Replace with your DataFrame's text column
embeddings = get_embeddings_in_batches(texts, batch_size=32)
```

```python
import optuna
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score

# Assuming X_train, X_test, y_train, y_test are defined

def objective(trial):
    # Hyperparameter suggestions
    param = {
        'verbosity': 0,
        'objective': 'reg:absoluteerror',
        'booster': trial.suggest_categorical('booster', ['gbtree', 'gblinear', 'dart']),
        'lambda': trial.suggest_loguniform('lambda', 1e-5, 10.0),  # L2 regularization
        'alpha': trial.suggest_loguniform('alpha', 1e-5, 10.0),    # L1 regularization
        'max_depth': trial.suggest_int('max_depth',6, 25),
        'eta': trial.suggest_float('eta', 1e-4, 0.05, log=True),  # Learning rate
        'subsample': trial.suggest_float('subsample', 0.5, 1.0),
        'colsample_bytree': trial.suggest_float('colsample_bytree', 0.5, 1.0),
        'n_estimators': 100,
        'n_jobs': -1
    }
    
    model = xgb.XGBRegressor(**param)
    model.fit(X_train, y_train)
    
    preds = model.predict(X_test)
    mae = mean_absolute_error(y_test, preds)
    
    return mae  # Objective to minimize

# Create a study and optimize
study = optuna.create_study()
study.optimize(objective, n_trials=500)

# Best hyperparameters
print("Best hyperparameters: ", study.best_params)
print("Best MAE: ", study.best_value)

```
